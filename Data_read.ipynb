```python
#Converting months into seasons
def conv_seasons(month):
    if(month>=3) and (month<=5):
        return 'Spring'
    elif(month>=6) and (month<=8):
        return 'Summer'
    elif(month>=9) and (month<=11):
        return 'Autumn'
    else:
        return 'Winter'
```

```python
# Assigning datetime as weekday col (Monday = 0, Sunday = 6)
td_clean['weekday'] = td_clean['started_at'].dt.dayofweek
```

```python
#Caluclating trip duration in minutes
td_clean['trip_duration_minutes'] = round((td_clean['ended_at']-td_clean['started_at']).dt.total_seconds()/60,2)
```

```python
# Extracting season
td_clean['season'] = td_clean['started_at'].dt.month.map(conv_seasons)
```
